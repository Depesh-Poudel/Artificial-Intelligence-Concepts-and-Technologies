<h2>Problem - 4: Matrix Operations</h2>
<p>
    For the following arrays:
</p>
<pre>
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
</pre>
<p>
    Prove the following with NumPy:
</p>
<ol>
    <li>Prove <code>A.A⁻¹ = I</code>.</li>
    <li>Prove <code>AB ≠ BA</code>.</li>
    <li>Prove <code>(AB)<sup>T</sup> = B<sup>T</sup>A<sup>T</sup></code>.</li>
</ol>
<p>
    Solve the following system of linear equations using inverse methods:
</p>
<pre>
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9
</pre>
<ol>
    <li>
        Represent the equations in matrix form using a NumPy array.
    </li>
    <li>
        Solve for <code>AX = B</code>.
    </li>
    <li>
        Solve the equations using <code>np.linalg.inv</code> function. <br>
        <b>Hint:</b> Explore more about the <code>linalg</code> module in NumPy.
    </li>
</ol>

```python
# Define matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A.A⁻¹ = I
A_inverse = np.linalg.inv(A)       # Inverse of A
identity_matrix = np.dot(A, A_inverse)
print("Qn 1 :-")
print("A.A⁻¹ = I:\n", identity_matrix)
print("Prove A.A⁻¹ = I: ", np.array_equal(identity_matrix, np.eye(2)))
print("-----------------------------------------")

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nQn 2 :-")
print("AB:\n", AB)
print("BA:\n", BA)
print("Prove AB ≠ BA: ", not np.array_equal(AB, BA))
print("-----------------------------------------")

# 3. Prove (AB)ᵀ = BᵀAᵀ
AB_T = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\nQn 3 :-")
print("(AB)ᵀ:\n", AB_T)
print("BᵀAᵀ:\n", BT_AT)
print("Prove (AB)ᵀ = BᵀAᵀ: ", np.array_equal(AB_T, BT_AT))
print("-----------------------------------------")

# Solve the linear system using inverse methods
# System of equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent the system as AX = B
coeff_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
rhs_vector = np.array([-1, -3, 9])

# Solve for X using np.linalg.inv
coeff_matrix_inv = np.linalg.inv(coeff_matrix)
solution = np.dot(coeff_matrix_inv, rhs_vector)
print("\nSolution to the system using inverse methods:\n", solution)
print("-----------------------------------------")

# Alternative: Use np.linalg.solve (more efficient)
solution_alt = np.linalg.solve(coeff_matrix, rhs_vector)
print("\nSolution to the system using np.linalg.solve function:\n", solution_alt)
print("-----------------------------------------")
```
